3.3. Are there more dinners or lunches? Create a dataframe with this data. Are
there more smokers during lunches or dinners? Create another dataframe with
this data. Join the two dataframes by time of day and calculate the percent of
smokers at lunch and dinner. Compare the results.

```python
#more dinners or lunches?
lunch_df = tips_df['time'] == 'Lunch' #boolean mask for getting true false values. True for lunch false for dinner
lunch_df = tips_df[lunch_df] # Stores boolean mask information in lunch_df variable

dinner_df = tips_df['time'] == 'Dinner'
dinner_df = tips_df[dinner_df]
```

```python
lunch_count = lunch_df['time'].count() #Gets count of lunches
dinner_count = dinner_df['time'].count() #Gets count of dinners
time_df = pd.DataFrame({
    'time': ['Lunch', 'Dinner'],
    'count': [lunch_count, dinner_count]
})
```

There are more dinners (176) to lunches (68). This is stored in the time_df. 

```python
#more smokers during lunch or dinner
#smokers during lunch:
lunch_smoker_df = lunch_df['smoker'] == 'Yes'
lunch_smoker_df = lunch_df[lunch_smoker_df]
lunch_smoker_count = lunch_smoker_df['smoker'].count()
```

```python
#smokers during dinner
dinner_smoker_df = dinner_df['smoker'] == 'Yes'
dinner_smoker_df = dinner_df[dinner_smoker_df]
dinner_smoker_count = dinner_smoker_df['smoker'].count()
```

```python
smoker_df = pd.DataFrame({
    'time': ['Lunch', 'Dinner'],
    'count': [lunch_smoker_count, dinner_smoker_count]
})
```

```python
merged = pd.merge(smoker_df, time_df, on='time')
merged.columns = ['time', 'smokers', 'total']
merged
```

```python
lunch_smokers = merged.iloc[0,1] / merged.iloc[0,2]
print(f'Lunch smokers: {lunch_smokers:.2f}')
dinner_smokers = merged.iloc[1,1] / merged.iloc[1,2]
print(f'Dinner smokers: {dinner_smokers:.2f}')
```

4.1. Import the dataset file into a Pandas dataframe and identify the count of
missing values per column. Handle missing values based on column type and
explain your reasons behind selecting appropriate techniques. 

```python
df = pd.read_csv('C:/Users/samsc/Desktop/ADS-500B/Module3_Resources/Excel/avocado.csv')
```

```python
df.isna().sum()
```

```python
df['Date'].dropna() #For date column I decided to drop all null value rows because the date
#is important information especially because produce only lasts for so long before it spoils.

#For these columns I wanted to impute with the average median of each row. I decided to imputed instead of drop
#fill with zeros because the data in these categories are float values and I chose median because it is the central
#measure of tendency least affected by outliers
df['AveragePrice'].fillna(df['AveragePrice'].median())
df['TotalVolume'].fillna(df['TotalVolume'].median())
df['Small'].fillna(df['Small'].median())
df['Large'].fillna(df['Large'].median())
df['AllSizes'].fillna(df['AllSizes'].median())
df['TotalBags'].fillna(df['TotalBags'].median())

df['Type'].fillna('unknown') #categorical data filled with unknown
df['Year'].fillna('unknown')
df['Region'].fillna('unknown')
```

4.2. Convert the fields Type, Year and Region to categorical data type and
subset the dataframe to exclude region “TotalUS” and “West” and sort the
dataframe by date in ascending order. Is the mean price of an avocado higher in
2017 compared to 2016? 

```python
sub_df = df[~df['Region'].isin(['TotalUS', 'West'])] #Microsoft Copilot help with ~ mark and the isin method
sub_df['Date'].sort_values()

year2017 = sub_df['Year'] == 2017
year2017 = sub_df[year2017]
print(f'2017 average avocado price: {year2017['AveragePrice'].mean():.2f}')
year2016 = sub_df['Year'] == 2016
year2016 = sub_df[year2016]
print(f'2016 average avocado price: {year2016['AveragePrice'].mean():.2f}')
```

Yes the mean price for an avocado is slightly higher average price in 2017 than in 2016 by about 18 cents.
