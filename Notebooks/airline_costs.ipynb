1. Use a linear regression model to predict the number of customers each airline serves
from its length of flight and daily flight time per plane.

```python
# y = airline_costs.iloc[:, 4] #target variables
# X = airline_costs.iloc[:, [1, 3]] #predictor variables
# X = sm.add_constant(X)
# print(type(y[1]))
# a=airline_costs.iloc[:,1]
# b=airline_costs.iloc[:,3]
# print(type(a.iloc[1]))
# print(type(b.iloc[1]))
# print(X)
y = airline_costs["CustomersServed"]
X = airline_costs[["FlightLength", "DailyFlightTime"]]
```

```python
#create training and test data to predict accuracy of linear model
```

```python
X_train = X.iloc[0:22]
X_test = X.iloc[22:]
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)
y_train = y.iloc[0:22]
y_test = y.iloc[22:]
```

```python
model = sm.OLS(y_train, X_train).fit()
y_pred = model.predict(X_test)
print(y_pred) #prediction
print(y_test) #actual
print(model.summary()) #not a very accurate model 
```

Because the model had an R-squared value of 0.537, the model did not accurately predict the number of customers served. Some of the values from y_pred had customersserved in the negatives and it is not possible to serve negative amount of customers. With that in mind there might be a few reasons as to why this model is weak. The predictors are not very strong at showing how many customers are served, which is quite possible or that there is not enough training and test data to prove that these are strong predictors for having customers on airline flights.

2. Build another regression model to predict the total assets of an airline from the customers served by the airline.

```python
y=airline_costs['TotalAssets']
X=airline_costs['CustomersServed']
X_train = X.iloc[0:22]
X_test = X.iloc[22:]
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)
y_train = y.iloc[0:22]
y_test = y.iloc[22:]
```

```python
model = sm.OLS(y_train, X_train).fit()
y_pred = model.predict(X_test)
print(y_pred)
print(y_test)
print(model.summary()) #not a very accurate model 
```

```python
#The R-squared value of this model is higher and it is about 0.73. The correlation between customers served and total assets is stronger than the 
#previous regression model. This makes sense because customers spend money on the flights which gives an airline income for assets.
#That being said doesn't neccessarily mean it is a better model. The notes of the second model suggests that there is high multicollinearity 
#which means that a lot of the factors used are redundant, which could possibly mean there is overfitting. Another strong argument that this model is 
#overfitting is that there are not a lot of data in this dataset so lower data amount means one is more prone to overfitting. Both models had 
#CustomersServed as its column, once as a predictor and the other time as a target variable. I believe that since both models did not perform great,
#the first being highly inaccurate and the second having a strong argument for overfitting, that CustomersServed might not be the best choice for figuring
#out airline costs. I think a variable using Revenue as a target variable would produce a much better model. 
```

```python
y=airline_costs['Revenue']
X=airline_costs[['Investments', 'TotalOperatingCost', 'TotalAssets']]
X_train = X.iloc[0:22]
X_test = X.iloc[22:]
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)
y_train = y.iloc[0:22]
y_test = y.iloc[22:]
```

```python
model = sm.OLS(y_train, X_train).fit()
y_pred = model.predict(X_test)
print(y_pred)
print(y_test) 
print(model.summary()) #not a very accurate model 
```
