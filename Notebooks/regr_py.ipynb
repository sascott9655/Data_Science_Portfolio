```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv('C:/Users/samsc/Desktop/ADS-500B/Data_and_Code2/OA 8.1 - regression.csv')
```

```python
plt.scatter(df['x'], df['y']);
```

Assume line is y=mx+b (linear)

```python
y=df['y'] #dependent var
X=df['x'] #independent var
X = sm.add_constant(X) #adds constant value to our linear regression equation
```

```python
lr_model = sm.OLS(y, X).fit()
print(lr_model.summary())
```

Intercept     x 
-10.263      3.97

```python
X_prime = np.linspace(X.x.min(), X.x.max(), 10) #Uses data in our data set to create acccurate line
X_prime = sm.add_constant(X_prime) # Add 1.0 constants 
y_hat = lr_model.predict(X_prime) #Have our y value be predicted or based on the X values
```

```python
plt.scatter(X.x,y)
plt.plot(X_prime, y_hat, 'red');
```
